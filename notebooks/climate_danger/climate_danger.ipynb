```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../ref_espace_communes.csv')
df_weather = pd.read_csv('../weather/data_risk_weather.csv')
df_fire = pd.read_csv('../fire/data_risk_fire.csv')
df_pollution = pd.read_csv('data_pollution_risk.csv', index_col=0)
df.head()
```

```python
df['department'] = df['code_commune_INSEE'].astype(str).str[:2]
print(df.shape)
df.head()
```

```python
years = list(range(2014, 2025))  

df_years = pd.DataFrame({'year': years})

df = df.merge(df_years, how='cross')
print(df.shape)
df.head()
```

```python
df.duplicated().sum()
```

```python
df_weather['code_commune_INSEE'] = df_weather['code_commune_INSEE'].astype(str).str.zfill(5)
df_weather['department'] = df_weather['department'].astype(str).str.zfill(2)
print(df_weather.shape)
df_weather.head()
```

```python
df_merged = (
    df
    .merge(df_weather, 
           left_on=['code_commune_INSEE', 'department', 'year'],
           right_on=['code_commune_INSEE', 'department', 'year'],
           how='left')        
)

print(df_merged.shape)
df_merged.head()
```

```python
df_merged['code_commune_INSEE'].duplicated().value_counts()
```

```python
df_fire.rename(columns= {'INSEE code': 'code_commune_INSEE'}, inplace=True)
df_fire.head()
```

```python
df_merged = (
    df_merged
    .merge(df_fire,
           left_on=['code_commune_INSEE', 'department', 'year'],
           right_on=['code_commune_INSEE', 'department', 'year'], 
           how='left')
)
print(df_merged.shape)
df_merged.head()
```

```python
df_pollution = pd.read_csv('data_pollution_risk.csv', index_col=0)
df_pollution = df_pollution[df_pollution['code_commune_INSEE'] != '00000']
df_pollution = df_pollution.drop_duplicates()
```

```python
df_pollution = (
    df_pollution
    .groupby(
        ['year', 
         'code_commune_INSEE',
         'code_departement',
         ])['ATMO_quarter']
    .max()
    .reset_index()
)
```

```python
df_pollution['code_departement'].nunique()
```

```python
df_pollution['code_commune_INSEE'] = df_pollution['code_commune_INSEE'].drop_duplicates()
```

```python
print(df_pollution.shape)
df_pollution.duplicated().value_counts()
```

```python
df_pollution['ATMO_risk'] = (df_pollution['ATMO_quarter'] * 4) / 5
df_pollution.rename(columns={'code_departement': 'department'}, inplace=True)
df_pollution.head()
```

```python
df_pollution.drop(columns='ATMO_quarter', inplace=True)
```

```python
print(df_pollution.shape)
df_pollution.duplicated().value_counts()
```

```python
df_merged = (
    df_merged
    .merge(df_pollution, 
           left_on=['code_commune_INSEE', 'department', 'year'],
           right_on=['code_commune_INSEE', 'department', 'year'],
           how='left')
)
print(df_merged.shape)
df_merged.head()
```

```python
df_merged.duplicated().sum()
```

```python
df_merged.columns
```

```python
columns = [
    'code_postal', 
    'nom_commune_complet', 
    'nom_region', 
    'latitude', 
    'longitude'
]

df_merged.drop(columns=columns, inplace=True)
df_merged.head()
```

```python
df_merged[['fire_score', 'ATMO_risk']] = df_merged[['fire_score', 'ATMO_risk']].fillna(0)
df_merged.head()
```

```python
df_merged.isna().sum() / df_merged.shape[0]
```

```python
df_merged['department'].nunique()
```

```python
df_merged = df_merged[~df_merged['nom_departement'].str.contains('Corse', na=False)]
```

```python
df_merged['department'].nunique()
```

```python
df_merged.shape
```

```python
risk_cols = ['drought_score', 'heat_score', 'rainfall_score', 'extrem_events_score']

for col in risk_cols:
    df_merged[col] = df_merged.groupby(['department', 'year'])[col].transform(
        lambda x: x.fillna(x.mean())
    )
```

```python
columns = ['drought_score', 'heat_score', 'rainfall_score', 'extrem_events_score', 'fire_score']
df_merged['avg_risk_score'] = df_merged[columns].mean(axis=1)
```
